## 1.3 Make 5 Different Visualization

```python
#Define Labels
labels = 'Gas', 'Diesel'
sizes = [185, 20]
explode = (0, 0.1)

#Visualization
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Proprotion Of Fuel',fontsize=15, bbox={'facecolor':'0.8', 'pad':1})
plt.show()
```

```python
#Top 5 Automaker By Mpg
data_n = data.groupby(["make"]).mean()
sns.barplot(x="city-mpg", y=data_n.index, data=data_n, order= data_n.sort_values('city-mpg',ascending = False).index[:5])
plt.title('Top 5 Automaker by MPG', fontsize=15)
plt.xlabel('MPG', fontsize=10)
plt.ylabel('Automkaer', fontsize=10)
plt.show()
```

```python
#Visualize Scatter Plot
sns.scatterplot(x="engine-size", y="city-mpg", hue="fuel-type", data=data)
plt.title('Correlation Between City-Mpg VS Engine Size', fontsize=15)
plt.xlabel('Engine Size', fontsize=10)
plt.ylabel('City-Mpg', fontsize=10)
plt.show()
```

```python
#Visualize Boxplot
data = data[data['price'] != '?'].astype({'price': 'int'}) #drop the "?" and covert to numeric
sns.boxplot(x="body-style", y="price", data=data)
plt.title('Box Plot of Price By Body Style', fontsize=15)
plt.xlabel('Body Style', fontsize=10)
plt.ylabel('{Price}', fontsize=10)
plt.show()
```

## 2.1 Car brand with highest price

```python
#Car Brand With Highest Price
data_s = data.groupby('make').max()['price']
data_s.sort_values(ascending=False)[:1]
```

## 2.2 Maximum Horsepower for car with 6 cylinder engine

```python
data = data[data['horsepower'] != '?'].astype({'horsepower': 'int64'}) #drop the "?" and covert to numeric
```

```python
data_h = data.groupby('num-of-cylinders').max()['horsepower']
data_h[data_h.index == 'six']
```

## 2.3 Average peak-rpm for 'turbo' style aspiration car

```python
data = data.astype({'peak-rpm': 'int64'}) #covert to numeric
```

```python
data_p = data.groupby('aspiration').mean()['peak-rpm']
data_p[data_p.index == 'turbo']
```
